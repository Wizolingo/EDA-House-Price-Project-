### MUSIC AND MENTAL HEALTH (Exploratory Data Analysis)

### Context

As stated by the WHO constitution: "Health is a state of complete physical, mental, and social well-being and not merely the absence of disease or infirmity. This impies that mental health is an integral and essential component of health.

Mental health is determined by many factors ranging from socioeconomic, biological to environmental factors.

Music therapy, or MT, is the use of music to improve an individual's stress, mood, and overall mental health. MT is also recognized as an evidence-based practice, using music as a catalyst for "happy" hormones such as oxytocin.

Listening to music can reduce symptoms of anxiety and depression, and can help to promote feelings of relaxation and calmness. Nevertheless, not all genres of music have the same effects on people.

This case study aims to Identify if there exists any correlation between an individual's music taste and their self-reported mental condition.

### Removing the rows with unrealistic values

#### Examining the BPM column

```python
data[['BPM']].sort_values('BPM',ascending=False)
```

We will drop the rows with values 624 and 999999999.0 as these are not realistic BPM numbers (no music out there has a BPM that high)

```python
data.drop(data[data['BPM']>=624].index, inplace=True)
```

```python
data.shape
```

#### Examining the Hours per day column

```python
data[['Hours per day']].sort_values('Hours per day',ascending=False)
```

We will drop the rows with values 24 as it is not logical that someone listens to music all day long nonstop

```python
data.drop(data[data['Hours per day']==24].index, inplace=True)
```

```python
data.shape
```

### Handling missing values

Let's check the percentage of missing values

```python
pd.DataFrame({'Count':data.isnull().sum()[data.isnull().sum()>0],'Percentage':(data.isnull().sum()[data.isnull().sum()>0]/data.shape[0])*100})
```

- BPM column has the highest missing values percentage of 14.64%
- Age, Primary streaming service and Composer have the same number of missing values 

Now let's replace missing values for the above columns with their meausres of central tendency (mean,mode,median) as the case may be.

```python
data['Age'] = data['Age'].fillna(round(data['Age'].mean(), 0))
data['Primary streaming service'] = data['Primary streaming service'].fillna(data['Primary streaming service'].mode()[0])
data['While working'] = data['While working'].fillna(data['While working'].mode()[0])
data['Instrumentalist'] = data['Instrumentalist'].fillna(data['Instrumentalist'].mode()[0])
data['Composer'] = data['Composer'].fillna(data['Composer'].mode()[0])
data['Foreign languages'] = data['Foreign languages'].fillna(data['Foreign languages'].mode()[0])
data['Music effects'] = data['Music effects'].fillna(data['Music effects'].mode()[0])
```

We will handle the BPM column differently

```python
data_no_bpm = data[data['BPM'].isnull() == True]
data_no_bpm
```

We will use the mean BPM of each unique 'Fav genre' to input the missing values

```python
data_no_bpm['Fav genre'].unique()
```

```python
no_bpm_genres = ['Country', 'Hip hop', 'Pop', 'Rock', 'Metal', 'Classical', 
                     'Video game music', 'R&B', 'Folk', 'Jazz', 'EDM', 'Gospel', 'K pop', 'Rap']
```

```python
for i in no_bpm_genres:
    data['BPM'] = data['BPM'].fillna(round(data[data['Fav genre'] == i]['BPM'].mean(), 0))
```

Let's check to be sure our missing value treatments worked

```python
data.isnull().sum()
```

There are no more rows with missing values

### Handling Outliers

Let's look at the distributions of the numeric columns

```python
numeric_columns = ['Age', 'Hours per day', 'BPM', 'Anxiety', 'Depression', 'Insomnia','OCD']
plt.figure(figsize=(15, 15))

for i, variable in enumerate(numeric_columns):
    plt.subplot(4, 4, i + 1)
    plt.boxplot(data[variable], whis=1.5)
    plt.tight_layout()
    plt.title(variable)

plt.show()
```

We can see that AGE, BPM and Hours Per day columns have many outliers, but here are a few things to consider.
1. People listen to music even in old age so we can't consider all ages above 43 as outliers.
2. Hours per day still has some high values, but people actually do listen to music for long hours.
3. There aren't many songs out there with BPMs more than 200

With respect to the above factors we remove outliers as follows:

```python
data = data.drop(data[data['Age']>70].index, axis=0)
```

```python
data = data.drop(data[data['Hours per day']>15].index, axis=0)
```

```python
data = data.drop(data[data['BPM']>200].index, axis=0)
```
